```python
file1 = "Data/Output_data/new_unemployment.csv"
unemployment_df = pd.read_csv(file1)
unemployment_df
```

```python
unemployment_df.columns
```

```python
unemployment_df=unemployment_df.rename(columns={'Unemp Rate 2010':"2010", 'Unemp Rate 2011':"2011",
       'Unemp Rate 2012':"2012", 'Unemp Rate 2013':"2013", 'Unemp Rate 2014':"2014",
       'Unemp Rate 2015':"2015", 'Unemp Rate 2016':"2016", 'Unemp Rate 2017':"2017",
       'Unemp Rate 2018':"2018"})
unemployment_df=unemployment_df.set_index("County/State")
unemployment_df
```

```python
def unemp_states(source):
    return source[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018']]

def birth_states(source):
    return source[[ '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018']]
```

```python
#Region I: Connecticut, Maine, Massachusetts, New Hampshire, Rhode Island, Vermont
#Region 1:CT, ME, MA, NH, RI, VT

Region1_birth=population_states.loc[(population_states["State"]=="US")|(population_states["State"]=="CT")|(population_states["State"]=="ME")|
                                (population_states["State"]=="MA")|(population_states["State"]=="NH")|(population_states["State"]=="RI")|
                                (population_states["State"]=="VT")]
                               

Region1_unemp=unemployment_df.loc[(unemployment_df["State"]=="US")|(unemployment_df["State"]=="CT")|(unemployment_df["State"]=="ME")|
                                  (unemployment_df["State"]=="MA")|(unemployment_df["State"]=="NH")|(unemployment_df["State"]=="RI")|
                                  (unemployment_df["State"]=="VT")]

Region1_unemp_rate = unemp_states(Region1_unemp)
Region1_birth_rate = birth_states(Region1_birth)
```

```python
Region1_birth_rate.T.plot.line(figsize=(15,5),marker='o')
plt.grid()
plt.title("Region 1: Birth Rate from 2011-2018", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Birth Rate (%)")
plt.savefig("Data/Output_data/BR_Region1")

Region1_unemp_rate.T.plot.line(figsize=(15,5),marker='o')
plt.title("Region 1: Unemployment Rate from 2010-2018", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Unemployment Rate (%)")
plt.grid()
plt.savefig("Data/Output_data/UR_Region1")
plt.show()
```
