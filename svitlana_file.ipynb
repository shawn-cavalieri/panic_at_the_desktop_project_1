```python
import pandas as pd
```

```python
file = "Data/PopulationPerStateandCounty.csv"
population = pd.read_csv(file)
population.head()
```

```python
population = population.drop(population.index[0])
population.columns
```

```python
new_df=population[["State","Area_Name","CENSUS_2010_POP", "POP_ESTIMATE_2011", 
                   "POP_ESTIMATE_2012", "POP_ESTIMATE_2013", "POP_ESTIMATE_2014",
                   "POP_ESTIMATE_2015", "POP_ESTIMATE_2016", "POP_ESTIMATE_2017","POP_ESTIMATE_2018"
                    ]]
new_df
```

```python
#no Puerto Rico
new_df=new_df.loc[new_df["State"]!="PR",:]
new_df
```

```python
new_df=new_df.rename(columns={"Area_Name":"Area Name", "CENSUS_2010_POP":"2010 Population", "POP_ESTIMATE_2011":"2011 Population",
                             "POP_ESTIMATE_2012":"2012 Population", "POP_ESTIMATE_2013":"2013 Population", "POP_ESTIMATE_2014":"2014 Population",
                             "POP_ESTIMATE_2015":"2015 Population", "POP_ESTIMATE_2016":"2016 Population", "POP_ESTIMATE_2017":"2017 Population",
                             "POP_ESTIMATE_2018":"2018 Population",} )
new_df
```
